```python
def encontrar_maximo(lista):
  valor_maximo = lista[0]
  for numero in lista:
    if numero > valor_maximo:
      valor_maximo = numero
  return valor_maximo
```

```python
def busca_binaria(lista,alvo):
  esquerda=0
  direita = len(lista) - 1
  while esquerda <= direita:
    meio = (esquerda + direita) // 2
    if lista[meio] == alvo:
      return meio
    elif lista[meio] < alvo:
            esquerda = meio + 1
    else:
      direita = meio - 1
  return -1
```
